```python
# Zadanie 1

import math
```

```python
def calculate(x):
    cos1 = math.cos(math.exp(x))
    ln1 = math.log((1+x) ** 2)
    sqrt1 = math.sqrt(math.exp(math.cos(x)) + (math.sin(math.pi * x)) ** 2)
    sqrt2 = math.sqrt(1/x) + math.cos(x ** 2)
    total = (cos1 + ln1 + sqrt1 + sqrt2) ** math.sin(x)
    return total
```

```python
x = 1.79
result = calculate(x)
result
```

```python
# Zadanie 2

def search_substr(subst, st):
    if subst.lower() in st.lower():
        return "Jest kontakt"
    else:
        return "Mimo!"
```

```python
search_substr("Hello", "Hello World")
```

```python
# Zadanie 3

from collections import Counter
```

```python
def top3(x):
    filtered = x.replace(' ', '')
    cnt = Counter(filtered)
    top = Counter(filtered).most_common(3)
    return top
```

```python
x = "fefebfheh fefefd hhrrhf d"
print(top3(x))
```

```python
#Zadanie 4

def count_it(sequence):
    digits = [int(i) for i in str(sequence)]
    counter = Counter(digits).most_common(3)
    return dict(counter)
```

```python
x = 122333444455555
print(count_it(x))
```

```python
#Zadanie 5

def queen(my_x, my_y, enemy_x, enemy_y):
    print("Is the queen dangerous for you?")
    if my_x == enemy_x or my_y == enemy_y:
        print("Yes")
    elif my_x - enemy_x == my_y - enemy_y:
        print("Yes")
    else:
        print("No")

def horse(my_x, my_y, enemy_x, enemy_y):
    dx = my_x - enemy_x
    dy = my_y - enemy_y
    print("Is the horse dangerous for you?")
    if (dx == 2 and dy == 1) or (dx == 1 and dy == 2):
        print("Yes")
    else:
        print("No")
```

```python
def game():
    my_x = int(input("Write your x: "))
    my_y = int(input("Write your y: "))
    enemy_x = int(input("Write your enemy's x: "))
    enemy_y = int(input("Write your enemy's y: "))
    if my_x < 1 or my_y < 1 or enemy_x < 1 or enemy_y < 1 or my_x > 8 or my_y > 8 or enemy_x > 8 or enemy_y > 8:
        print("Not possible")
        return False
    if (my_x, my_y) == (enemy_x, enemy_y):
        print("Not possible")
        return False
    queen(my_x, my_y, enemy_x, enemy_y)
    horse(my_x, my_y, enemy_x, enemy_y)
```

```python
game()
```
