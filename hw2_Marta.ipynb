```python
# Zadanie 1

import math   
```

```python
def solve(a,b,c):
    disc = b**2 - 4*a*c
    if disc == 0:
        sol = -b/2*a
        print(f"Solution: {sol}")
    elif disc < 0:
        print("No solution")
    else:
        sol_1 = (-b + math.sqrt(disc))/2*a
        sol_2 = (-b - math.sqrt(disc))/2*a
        print(f"First solution: {sol_1}, Second solution: {sol_2}")
```

```python
solve(1,-12,20)
solve(1,17,72)
solve(1,-7,-44)
solve(1,9,8)
solve(1,-2,-63)
solve(1,4,-5)
```

```python
# Zadanie 2

def reverse(s):
    reverse = s[::-1]
    return reverse
```

```python
# Zadanie 3

def change(lst):
    temp = lst[0::]
    lst[0] = temp[-1]
    lst[-1] = temp[0]
    return lst
```

```python
lst = [1, 2, 3, 4, 5]
change(lst)
```

```python
# Zadanie 4

def useless(s):
    temp = 0
    for i in s:
        if i > temp:
            temp = i
    return temp/len(s)
```

```python
s = [1, 5, 3, 4, 2]
useless(s)
```

```python
# Zadanie 5

def all_eq(lst):
    max_len = 0
    for i in lst:
        if int(len(str(i))) > max_len:
            max_len = int(len(str(i)))
    return [i + '_' * (max_len - len(s)) for i in lst]
```

```python
lst = ["крот", "белка", "выхухоль"]
all_eq(lst)
```

```python
# Zadanie 1.2

def compress(s):
    uniq = ""
    count = 1
    for i in range(1, len(s)):
        if s[i] == s[i-1]:
            count += 1
        else:
            uniq += s[i-1] + str(count)
            count = 1
    uniq += s[-1] + str(count)
    if len(uniq) > len(s):
        return s
    else:
        return uniq
```

```python
s = "aabccccaa"
compress(s)
```

```python
# Zadanie 3.2

def longest(s):
    long = ''
    for i in range(len(s)):
        for j in range(i, len(s)):
            sub = s[i:j+1]
            if sub == sub[::-1] and len(sub) > len(long):
                long = sub
    return long
```

```python
s = "haabaad"
longest(s)
```

```python
# Zadanie 6

def tournament(x):
    rounds = 0
    while len(x) > 1:
        new_x = []
        for i in range(0, len(x), 2):
            if i+1 < len(x):
                rounds += 1
                new_x.append(max(x[i], x[i+1]))
            else:
                new_x.append(x[i])
        x = new_x
    return f"Winner: {x[0]}, Rounds: {rounds}"
```

```python
x = [1,2,3,4,6]
tournament(x)
```

```python
# Zadanie 7

def wave_sort(s):
    for i in range(0, len(s)-1, 2):
        if s[i] < s[i + 1]:
            s[i], s[i+1] = s[i+1], s[i]
    for i in range(1, len(s)-1, 2):
        if s[i] > s[i + 1]:
            s[i], s[i+1] = s[i+1], s[i]
    return s
```

```python
s = [1,2,3,4,5]
wave_sort(s)
```

```python
# Zadanie 9

def double(lst_1, lst_2):
    res = []
    for i in lst_1:
        if i in lst_2:
            res.append(i)
    return res
```

```python
a = [1,2,2,3]
b = [2,2,3,4]
double(a, b)
```

```python
# Zadanie 10

def inversion(s):
    count = 0
    for i in range(len(s)):
        for j in range(i+1, len(s)):
            if s[i] > s[j]:
                count += 1
    return count
```

```python
s = [4,2,3,5,1]
inversion(s)
```
