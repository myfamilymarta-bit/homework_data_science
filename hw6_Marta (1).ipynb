```python
class Message(ABC):
    @abstractmethod
    def prepare_text(self, text: str):
        pass

    @abstractmethod
    def send(self, recipient: str, text: str):
        pass
        
    @staticmethod
    def generate_id():
        return int(time.time() * random.randint(1000,2000))

class EmailMessage(Message):
    def prepare_text(self, text: str):
        return f"{text}"

    def send(self, recipient: str, text: str):
        prepared = self.prepare_text(text)
        Logger.log(Message.generate_id(), "Email", recipient, prepared)

class SMSMessage(Message):
    def prepare_text(self, text: str):
        return text[:160]
        
    def send(self, recipient: str, text: str):
        prepared = self.prepare_text(text)
        Logger.log(Message.generate_id(), "SMS", recipient, prepared)
```

```python
class Logger:
    log_entries = []

    @staticmethod
    def log(msg_id: int, channel: str, recipient: str, text: str):   
        Logger.log_entries.append({
            "id": msg_id,
            "channel": channel,
            "recipient": recipient,
            "text": text
        })

    @staticmethod
    def show_log():  
        for zapis in Logger.log_entries:
            print(zapis)
```

```python
email = EmailMessage()
sms = SMSMessage()
email.send("user@example.com", "Hello, it's me")
sms.send("+123456789", "Cheburek")
Logger.show_log()
```

```python
# Zadanie 2

class Engine(ABC):
    @abstractmethod
    def start(self):
        pass

    @abstractmethod
    def stop(self):
        pass

class PetrolEngine(Engine):
    def start(self):
        print("Petrol engine is turned on")
    def stop(self):
        print("Petrol engine is turned off")

class ElectricEngine(Engine):
    def start(self):
        print("Electric engine is turned on")
    def stop(self):
        print("Electric engine is turned off")
```

```python
class Car:
    def __init__(self, vin: str, engine_type: str):
        self.vin = vin
        self._engine = None
        self._create_engine(engine_type)

    def _create_engine(self, engine_type: str):
        if not Car.validate_vin(self.vin):
            print (f"Invalid VIN: {vin}")
            return
        if engine_type == "petrol":
            self._engine = PetrolEngine()
        elif engine_type == "electric":
            self._engine = ElectricEngine()
        else:
             return "Unknown engine"

    def start(self):
        if self._engine:
            self._engine.start()
        else:
            print("No engine")

    def stop(self):
        if self._engine:
            self._engine.stop()

    @staticmethod
    def validate_vin(vin: str):
        if len(vin) != 17:
            return False
        return vin.isalnum() and vin.isupper()
```

```python
vin = "1G1FA1E34K1234567"
car = Car(vin, "petrol")
car.start()
car.stop()
```

```python
# Zadanie 3
class PaymentMethod(ABC):
    @abstractmethod
    def calculate_commission(self, amount: float):
        pass

    @abstractmethod
    def pay(self, amount: float):
        pass

class CreditCardPayment(PaymentMethod):
    def calculate_commission(self, amount: float):
        return amount * 0.02 

    def pay(self, amount: float):
        commission = self.calculate_commission(amount)
        print(f"Pay by card: {amount}, commission: {commission}")

class DigitalMoneyPayment(PaymentMethod):
    def calculate_commission(self, amount: float):
        return amount * 0.005 

    def pay(self, amount: float):
        commission = self.calculate_commission(amount)
        print(f"Pay with digital money: {amount}, commission: {commission}")
```
