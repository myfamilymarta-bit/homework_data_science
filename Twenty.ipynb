```python
# Zadanie 1

def octopus_typing(text):
    if not text:
        return ""
    result = []
    i = 0
    n = len(text)
    while i < n:
        used_letters = set() 
        digit_count = {}     
        used_symbols = set()
        round_chars = 0
        round_result = []
        while i < n and round_chars < 8:
            ch = text[i]
            if ch.isalpha(): 
                if ch.lower() in used_letters:
                    round_result.append('*')
                else:
                    round_result.append(ch)
                    used_letters.add(ch.lower())
            elif ch.isdigit(): 
                count = digit_count.get(ch, 0)
                if count >= 2:
                    round_result.append('*')
                else:
                    round_result.append(ch)
                    digit_count[ch] = count + 1
            else: 
                if ch in used_symbols:
                    round_result.append('*')
                else:
                    round_result.append(ch)
                    used_symbols.add(ch)
            round_chars += 1
            i += 1
        result.extend(round_result)
    return ''.join(result)
```

```python
print(octopus_typing("Letter"))  
print(octopus_typing("Unconcerned"))  
print(octopus_typing("Can a can contain fish?"))  
print(octopus_typing("Write 122 pages")) 
print(octopus_typing("8000 dollars went missing"))  
print(octopus_typing("")) 
```

```python
# Zadanie 2

def string_expansion(s):
    if not s:
        return ""
    result = []
    multiplier = 1 
    i = 0
    while i < len(s):
        if s[i].isdigit():
            multiplier = int(s[i])
            i += 1
        else:
            result.append(s[i] * multiplier)
            i += 1
    return ''.join(result)
```

```python
print(string_expansion("3D2a5d2f"))
print(string_expansion("3abc"))      
print(string_expansion("3d332f2a"))  
print(string_expansion("abcde"))   
print(string_expansion("1111"))     
print(string_expansion(""))         
```

```python
# Zadanie 3

def prefill(n, v=None):
    if isinstance(n, str):
        if n and n[0] != '-' and n.isdigit():
            n = int(n)
        else:
            return "Error"
    else:
        if not isinstance(n, int) or isinstance(n, bool):
            return "Error"
    if n < 0:
        return "Error"
    return [v] * n
```

```python
print(prefill(3, 1))        
print(prefill(2, "abc"))   
print(prefill("1", 1))        
print(prefill(3, prefill(2, '2d'))) 
print(prefill("-123", 1))  
```

```python
# Zadanie 4

def solution(s):
    result = []
    for i in range(0, len(s), 2):
        pair = s[i:i+2]
        if len(pair) == 2:
            result.append(pair)
        else:
            result.append(pair + '_')
    return result
```

```python
print(solution('abc'))      
print(solution('abcdef'))   
print(solution(''))        
print(solution('a'))        
```

```python
def find_missing_letter(chars):
    if chars[0].islower():
        alphabet = string.ascii_lowercase
    else:
        alphabet = string.ascii_uppercase
    start_index = alphabet.index(chars[0])
    expected_letters = alphabet[start_index:]
    for i in range(len(expected_letters)):
        expected_char = expected_letters[i] 
        if i >= len(chars) or chars[i] != expected_char:
            return expected_char
```

```python
print(find_missing_letter(['a','b','c','d','f']))  
print(find_missing_letter(['O','Q','R','S']))     
print(find_missing_letter(['t','u','v','w','x','z']))  
print(find_missing_letter(['A','B','C','D','F']))  
```

```python
# Zadanie 6

def find_starters(record):
    all_people = set(record.keys())
    receivers = set()
    for told_list in record.values():
        receivers.update(told_list)
    starters = all_people - receivers
    return sorted(list(starters))
```

```python
record1 = {"Sarah": ["Elle"], "Larry":["Sarah","Elle"], "John":["Larry","May"], 
           "May":["Sarah"], "Elle":["May"]}
print(find_starters(record1)) 

record2 = {"Mike":["Paige","Rose"], "Paige":["Hugh"], "Liam":["Rose"],
           "Rose":["Paige", "Hugh"], "Hugh":["Liam"], "Victor":["Rose","Liam"]}
print(find_starters(record2))
```

```python
# Zadanie 7

def extract_value(element):
    if isinstance(element, int):
        return element
    else: 
        return element[0]

def sort_by_content(lst):
    return sorted(lst, key=extract_value)
```

```python
# Zadanie 8

def check_letters(txt, first, second):
    for i in range(len(txt) - 1):
        if txt[i] == first:
            if txt[i + 1] != second:
                return False
    if first == second and txt[-1] == first:
        return False
    return True
```

```python
print(check_letters("he headed to the store", "h", "e"))  
print(check_letters("abcdee", "e", "e"))                
print(check_letters("i found an ounce with my hound", "o", "u")) 
print(check_letters("we found your dynamite", "d", "y"))   
```

```python
# Zadanie 9

def complete_cycle(lst):
    visited = set()
    index = 0
    for i in range(len(lst)):
        if index in visited:
            return False
        visited.add(index)
        index = lst[index]
    return index == 0  
```

```python
print(complete_cycle([1, 4, 3, 0, 2])) 
print(complete_cycle([1, 4, 0, 3, 2])) 
print(complete_cycle([5, 3, 4, 2, 0, 1])) 
```

```python
# Zadanie 10

def find_series_type(arr):
    n = len(arr)
    if n <= 1:
        return 2
    if all(x == 0 for x in arr):
        return 0
    diff = arr[1] - arr[0]
    is_ap = True
    for i in range(n - 1):
        if arr[i + 1] - arr[i] != diff:
            is_ap = False
            break
    is_gp = True
    if arr[0] == 0:
        is_gp = False
    else:
        for i in range(n - 1):
            if arr[i] * arr[1] != arr[0] * arr[i + 1]:
                is_gp = False
                break
    if is_ap and is_gp:
        return 2
    elif is_ap:
        return 0
    elif is_gp:
        return 1
    else:
        return -1
```

```python
print(find_series_type([2, 5, 8, 11, 14])) 
print(find_series_type([1, 2, 4, 8, 16]))   
print(find_series_type([1, 2, 1, 3, 4, 5])) 
print(find_series_type([1, 1, 1, 1, 1]))    
print(find_series_type([0, 0, 0, 0, 0]))    
print(find_series_type([1, 2]))          
print(find_series_type([0, 0]))            
print(find_series_type([0, 1]))          
print(find_series_type([1, 0]))          
```

```python
# Zadanie 11

def differences(lst):
    while len(lst) > 1:
        lst = [abs(lst[i] - lst[i + 1]) for i in range(len(lst) - 1)]
    return lst[0]
```

```python
print(differences([1, 2, 3]))             
print(differences([1, 5, 2, 7, 8, 9, 0]))  
print(differences([2, 3, 1]))           
```

```python
# Zadanie 12

def mirror(arr):
    sorted_arr = sorted(arr)
    return sorted_arr + sorted_arr[-2::-1]
```

```python
# Zadanie 13

def add_no_carry(a, b):
    a_str = str(a)
    b_str = str(b)
    result_parts = []
    if len(a_str) < len(b_str):
        a_str = '0' * (len(b_str) - len(a_str)) + a_str
    elif len(b_str) < len(a_str):
        b_str = '0' * (len(a_str) - len(b_str)) + b_str
    for i in range(len(a_str)):
        digit_sum = int(a_str[i]) + int(b_str[i])
        result_parts.append(str(digit_sum))
    return int(''.join(result_parts))
```

```python
print(add_no_carry(248, 208))  
print(add_no_carry(6, 9))      
print(add_no_carry(1, 1))      
print(add_no_carry(9, 11))     
print(add_no_carry(14, 0))     
```

```python
# Zadanie 14

def infected(s):
    continents = s.split('X')
    total = 0
    infected_count = 0
    for cont in continents:
        if not cont:  
            continue
        total += len(cont)
        if '1' in cont:
            infected_count += len(cont)
    if total == 0:
        return 0.0
    return 100.0 * infected_count / total
```

```python
print(infected("01000000X000X011X0X")) 
print(infected("X"))                     
print(infected("0X1"))                   
print(infected("111"))                  
```

```python
# Zadanie 17

def count_valid_binary_strings(n):
    if n == 1:
        return 1
    a, b = 1, 1 
    for i in range(1, n):
        a, b = b, a + b
    return b
```

```python
print(count_valid_binary_strings(1))  
print(count_valid_binary_strings(2)) 
print(count_valid_binary_strings(3))  
```

```python
# Zadanie 18

def pop_blocks(lst):
    if not lst:
        return []
    while True:
        n = len(lst)
        i = 0
        removed = False
        while i < n:
            j = i
            while j < n and lst[j] == lst[i]:
                j += 1
            run_length = j - i
            if run_length > 1:
                lst = lst[:i] + lst[j:]
                removed = True
                break
            i = j
        if not removed:
            break
    return lst
```

```python
print(pop_blocks(["A", "B", "C", "C", "B", "D", "A"]))  
print(pop_blocks(["B", "B", "A", "C", "A", "A", "C"])) 
print(pop_blocks(["B", "B", "C", "C", "A", "A", "A"]))  
print(pop_blocks(["C", "A", "C"]))  
print(pop_blocks(['ab', 'ab', 'cd', 'cx', 'B']))
```

```python
# Zadanie 20

def robot_instructions(path):
    if not path:
        return "Paused"
    directions = {
        '^': 'up',
        'v': 'down',
        '>': 'right',
        '<': 'left'
    }
    result = []
    count = 1
    current = path[0]
    for i in range(1, len(path)):
        if path[i] == current:
            count += 1
        else:
            result.append(f"Take {count} step {directions[current]}")
            current = path[i]
            count = 1
    result.append(f"Take {count} step {directions[current]}")
    return '\n'.join(result)
```

```python
print(robot_instructions("^^vvvv>><<^v>"))
print(robot_instructions("")) 
print(robot_instructions("^^^")) 
print(robot_instructions("><"))  
```
