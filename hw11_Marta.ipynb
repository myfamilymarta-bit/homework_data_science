```python
import numpy as np
```

```python
def solve(a,b,c):
  a_np = np.array(a)
  b_np = np.array(b)
  c_np = np.array(c)
  print(f"Addition A+B:\n {a_np + b_np}")
  print(f"Subtraction B-A:\n {b_np - a_np}")
  print(f"Matrix Multiplication A*C:\n {a_np @ c_np}")
  print(f"Matrix Multiplication A*B*C:\n {a_np @ b_np @ c_np}")
```

```python
a = [[1, 2, 4], [7, 5, -2], [-2, 1, 3]]
b = [[-1, 0, 3], [2, -6, 1], [-7, 1, 5]]
c = [[-2, 3], [4, 0], [7, -1]]
solve(a,b,c)
```

```python
a = [[0, 3, 4], [-7, 1, -2], [2, -1, 3]]
b = [[1, 2, 4], [7, 0, -2], [-3, -2, 5]]
c = [[-3, 2], [0, 4], [-1, 7]]
solve(a,b,c)
```

```python
a = [[5, 2, 9], [7, 0, -2], [-2, 1, 6]]
b = [[1, -2, 1], [7, 5, -2], [-2, 8, 3]]
c = [[-2, 3], [4, 5], [0, -1]]
solve(a,b,c)
```

```python
a = [[1, -2, 8], [7, 5, -2], [-2, 3, 0]]
b = [[0, 2, 4], [1, -3, -2], [-2, 1, 3]]
c = [[0, 3], [5, -4], [7, -1]]
solve(a,b,c)
```

```python
a = [[9, 2, 4], [7, 5, -5], [6, -2, 3]]
b = [[8, 2, 4], [0, 1, -3], [-2, 1, 6]]
c = [[-2, 0], [4, 8], [1, -1]]
solve(a,b,c)
```

```python
def reverse(a):
  np_a = np.array(a)
  print(f"Original Matrix:\n {np_a}")
  print(f"Reversed Matrix:\n {np.linalg.inv(np_a)}")
```

```python
a = [[1, -2, 8], [7, 5, -2], [-2, 3, 0]]
reverse(a)
```

```python
a = [[5, 2, 9], [7, 0, -2], [-2, 1, 6]]
reverse(a)
```
