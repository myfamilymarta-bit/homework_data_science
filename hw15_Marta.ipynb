```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

```python
import kagglehub
from kagglehub import KaggleDatasetAdapter
```

```python
file_path = "creditcard.csv"

df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "mlg-ulb/creditcardfraud",
  file_path,
)

print("First 5 records:", df.head())
print("Description:", df.describe())
```

```python
X = df.drop("Class", axis=1)
y = df['Class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=65)
```

```python
clf = LogisticRegression(random_state=0, max_iter=1000).fit(x_train, y_train)
print(f"Training score: {clf.score(x_train, y_train)}")
```

```python
y_pred = clf.predict(x_test)
```

```python
print(f"Precision: {precision_score(y_test, y_pred)}")
print(f"Recall: {recall_score(y_test, y_pred)}")
print(f"F1_score: {f1_score(y_test, y_pred)}")
```

# Логистическая регрессия

```python
df = pd.read_csv("train_data.csv")
```

```python
X = df.drop("Survived", axis=1)
y = df['Survived']
```

```python
mean = np.mean(X, axis=0)
std = np.std(X, axis=0)
X = (X - mean) / std
```

```python
a = np.zeros(X.shape[1])
b = 0.0

for i in range(1000):
    linear_model = np.dot(X, a) + b
    y_pred = 1 / (1 + np.exp(-linear_model))

    a -= 0.1 * np.dot(X.T, (y_pred - y)) / len(y)
    b -= 0.1 * np.mean(y_pred - y)

final_linear_model = np.dot(X, a) + b
y_pred_class = (1 / (1 + np.exp(-final_linear_model)) >= 0.5).astype(int)

print(f"Precision: {precision_score(y, y_pred_class)}")
print(f"Recall: {recall_score(y, y_pred_class)}")
print(f"F1-score: {f1_score(y, y_pred_class)}")
```
