```python
from abc import ABC, abstractmethod
```

```python
# Zadanie 1: Factory

class Transport(ABC):
    @abstractmethod
    def deliver(self):
        pass

class Truck(Transport):
    def deliver(self):
        return "Delivery on land"

class Ship(Transport):
    def deliver(self):
        return "Delivery on sea"

class Plane(Transport):
    def deliver(self):
        return "Delivery by air"
```

```python
class Logistics:
    @staticmethod
    def create_transport(way):
        if way == "road":
            return Truck()
        elif way == "sea":
            return Ship()
        elif way == "air":
            return Plane()
```

```python
transport = Logistics.create_transport("road")
print(transport.deliver())

transport = Logistics.create_transport("sea")
print(transport.deliver())

transport = Logistics.create_transport("air")
print(transport.deliver())
```

```python
# Zadanie 2: Abstract Factory

class Button(ABC):
    @abstractmethod
    def render(self):
        pass

class Checkbox(ABC):
    @abstractmethod
    def render(self):
        pass
```

```python
class ABCFactory:
    @staticmethod
    def create_factory(os_type):
        if os_type == "windows":
            return WindowsFactory()
        elif os_type == "mac":
            return MacFactory()

class WindowsFactory:
    def create_button(self):
        return "Windows style button"
    def create_checkbox(self):
        return "Windows style checkbox"

class MacFactory:
    def create_button(self):
        return "macOS style button"
    def create_checkbox(self):
        return "macOS style checkbox"
```

```python
factory = ABCFactory.create_factory("windows")
button = factory.create_button()
checkbox = factory.create_checkbox()
print(button)
print(checkbox)

factory = ABCFactory.create_factory("mac")
button = factory.create_button()
checkbox = factory.create_checkbox()
print(button)
print(checkbox)
```

```python
# Zadanie 3: Decorator

class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def get_salary(self):
        return self.salary

    def get_description(self):
        return f"{self.name}: base salary {self.salary}"

class ProductivityBonus:
    def __init__(self, employee):
        self.employee = employee

    def get_salary(self):
        return self.employee.get_salary() + 5000

    def get_description(self):
        return self.employee.get_description() + " + productivity bonus 5000"

class SeniorityBonus:
    def __init__(self, employee):
        self.employee = employee

    def get_salary(self):
        return self.employee.get_salary() + 3000

    def get_description(self):
        return self.employee.get_description() + " + seniority bonus 3000"

class ManagementBonus:
    def __init__(self, employee):
        self.employee = employee

    def get_salary(self):
        return self.employee.get_salary() + 10000

    def get_description(self):
        return self.employee.get_description() + " + management bonus 10000"
```

```python
emp = Employee("Marta", 50000)
print(emp.get_description())
print(f"Total: {emp.get_salary()}")

emp = ProductivityBonus(emp)
emp = SeniorityBonus(emp)
emp = ManagementBonus(emp)

print(emp.get_description())
print(f"Total: {emp.get_salary()}")
```
