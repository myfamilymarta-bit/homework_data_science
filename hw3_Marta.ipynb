```python
# Zadanie 3

def list_int(n):
    if n > 0:
        list_ = [i for i in range(1,n+1)]
        return list_[::-1]
    else:
        return "Error"
```

```python
n = 5
list_int(n)
```

```python
# Zadanie 5

def digit(num):
    num = [int(d) for d in str(num)[::-1]]
    return num
```

```python
num = 35231
digit(num)
```

```python
# Zadanie 6

def matrix_col(n):
    m = []
    for j in range(n):
        m.append([0] * n)
    num = 1
    for col in range(n):
        if col % 2 == 0:
            for row in range(n):
                m[row][col] = num
                num += 1
        else:
            for row in range(n-1, -1, -1):
                m[row][col] = num
                num += 1
    return m
    
```

```python
def matrix_row(n):
    m = []
    for j in range(n):
        m.append([0] * n)
    num = 1
    for row in range(n):
        if row % 2 == 0:
            for col in range(n):
                m[row][col] = num
                num += 1
        else:
            for col in range(n-1, -1, -1):
                m[row][col] = num
                num += 1
    return m
    
```

```python
n = 3
matrix_col(n)
```
