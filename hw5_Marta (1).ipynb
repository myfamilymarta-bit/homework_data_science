```python
class BankAccount:
    def __init__(self, byn: float = 0, usd: float = 0, eur: float = 0):
        self.byn = byn
        self.usd = usd
        self.eur = eur
        self.rates = {'BYN': 1.0, 'USD': 3.2, 'EUR': 3.5}

    def deposit(self, amount: float, currency: str):
        if amount <= 0:
            return
        if currency == 'BYN':
            self.byn += amount
        elif currency == 'USD':
            self.usd += amount
        elif currency == 'EUR':
            self.eur += amount

    def withdraw(self, amount: float, currency: str):
        if amount <= 0:
            return
        if currency == 'BYN' and self.byn < amount:
            print("Not enough money, loser")
            return
        if currency == 'USD' and self.usd < amount:
            print("Not enough money, loser")
            return
        if currency == 'EUR' and self.eur < amount:
            print("Not enough money, loser")
            return
        if currency == 'BYN':
            self.byn -= amount
        elif currency == 'USD':
            self.usd -= amount
        elif currency == 'EUR':
            self.eur -= amount

    def transfer(self, amount: float, from_currency: str, to_currency: str):
        if from_currency == to_currency:
            return
        if amount <= 0:
            return
        if from_currency == 'BYN' and self.byn < amount:
            print("Not enough money, loser")
            return
        if from_currency == 'USD' and self.usd < amount:
            print("Not enough money, loser")
            return
        if from_currency == 'EUR' and self.eur < amount:
            print("Not enough money, loser")
            return
        amount_in_byn = amount * self.rates[from_currency]
        amount_in_to = amount_in_byn / self.rates[to_currency]
        if from_currency == 'BYN':
            self.byn -= amount
        elif from_currency == 'USD':
            self.usd -= amount
        elif from_currency == 'EUR':
            self.eur -= amount
        if to_currency == 'BYN':
            self.byn += amount_in_to
        elif to_currency == 'USD':
            self.usd += amount_in_to
        elif to_currency == 'EUR':
            self.eur += amount_in_to

    def total_in_currency(self, target_currency: str):
        total_byn = (self.byn * self.rates['BYN'] + self.usd * self.rates['USD'] + self.eur * self.rates['EUR'])
        return total_byn / self.rates[target_currency]

    def write(self):
        return f"BYN: {self.byn:.2f}, USD: {self.usd:.2f}, EUR: {self.eur:.2f}"
```

```python
account = BankAccount()
account.deposit(100, "BYN")
account.deposit(50, "USD")
account.deposit(30, "EUR")
account.write()
```

```python
account.transfer(10, "USD", "EUR")
account.write()
```

```python
account.withdraw(20, "BYN")
account.write()
```

```python
total_byn = account.total_in_currency("BYN")
print(total_byn)
```

```python
total_usd = account.total_in_currency("USD")
print(f"{total_usd:.2f}")
```
