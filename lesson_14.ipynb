```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
```

```python
df_clean = df[['GrLivArea', 'OverallQual', 'YearBuilt', 'SalePrice']].dropna()

x = np.array(df_clean[['GrLivArea', 'OverallQual', 'YearBuilt']])
y = np.array(df_clean['SalePrice'])

x = x[:100]
y = y[:100]

X_mean = np.mean(x, axis=0)
X_std = np.std(x, axis=0)
X_scaled = (x - X_mean) / X_std
```

```python
a = np.zeros(3)
b = 0.0

learning_rate = 0.001
iterations = 1000

for i in range(iterations):
    y_pred = np.dot(X_scaled, a) + b

    error = np.mean((y_pred - y) ** 2)

    d_a = 2 * np.dot(X_scaled.T, (y_pred - y)) / len(y)
    d_b = 2 * np.mean(y_pred - y)

    a = a - learning_rate * d_a
    b = b - learning_rate * d_b

print(f"\nФинальные коэффициенты:")
print(f"a = {a}")
print(f"b = {b:.2f}")

print("\nПроверка на первых 5 точках:")
for i in range(5):
    pred_scaled = np.dot(X_scaled[i], a) + b
    print(f"x1={x[i,0]:.0f}, x2={x[i,1]:.0f}, x3={x[i,2]:.0f}, реальная={y[i]:.0f}, предсказанная={pred_scaled:.0f}")
```
